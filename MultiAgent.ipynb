### Multi-Agent System Architecture:

The older architectures where we used to do everything with a single agent handles all the task with all the tools binded to the llm due to which if our goal or the system was complex we used to face a lot of problems that used to degrade the efficientcy of our system. The problems were like:-

- An Agent getting binded with the multiple tools would create ambiguity for the llm to choose the better tool due to which it would creaate tools that were not efficient for the task or the goal

- In the complex agentic system, we need the specialized agents to perform the steps to reach the goal, which is not possible for an single agent to do all the specific tasks like researcher, math expertise, code review, code generator, validator, Agentic RAG, SocialMedia Analyzer etc to do a goal for a big agentic system

- for a single agent to achieve a goal, e had to track a lot of data and information in one state which could be complex to do as well as the memory can get conjusted (long term and short term memory)

To avoid such kind of conditions leading to inefficient system, we can use divide and conquer methods and use multiple autonomous agents that just need a goal for them to work autonomously with coordinationn with each other.
Each agents are speacialized in one task like:
Agent1 ---> Reasearcher

Agent2 ---> Math expertise

Agent3 ---> Q/A

Agent4 ---> Agentic RAG expert

Agent5 ---> code generator/evaluator etc

These independent agents can be anything from a single prompt and an LLM call to ReAct...Planner-Agent... or can contain whole sub-graphs within them.

Where, the Supervisor(orchestrator) agent would divide the goal into multiple small goals and route them to different agents that are required to reach the main goal. The supervisor agent also keeps the supervision on the agents and maintains the coordination, when the agents are finished then the supervisor will combine the results and submit.

Supervisor agent is the main planner and the system manager that protects our system from going into chaos, ambuguity and on the state where even the "Fault Tolerence" and "Human In The Loop" feature can't get us out from i.e. a complete deadlock.

We can say that the Supervisor is the controller of the Multi-AI Agent system.

Although we have smart, autunomous and specialized agents, they can only say what to do with the goal provided to it and not like when to do it. All that is done by one agent-----> Planner/ Supercisor/ Orchestrator Agent. Supervisr agent is the one that is gonna say what agents are gonna get goals and routed to.

The main goal of the multi-agent system is to make the system: 

- Modular (don't make the system conjusted)

- Specialized (distribute the work in the basis of the specialization)
